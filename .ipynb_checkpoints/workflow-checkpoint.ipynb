```python
train_split=int(0.8*len(X))
train_split
```

```python
X_train=X[:train_split]
Y_train=Y[:train_split]
X_test=X[train_split:]
Y_test=Y[train_split:]
```

```python
len(X_train),len(X_test)
```

```python
## Building a function to visualize
```

```python
len(X_train),X_test
```

```python
def plot_predictions(train_data=X_train,train_labels=Y_train,test_data=X_test,test_labels=Y_test,predictions=None):

    plt.figure(figsize=(10,7),dpi=1000)
    plt.scatter(train_data,train_labels,color='blue',label="training data")
    
    ## testing data
    plt.scatter(test_data,test_labels,color='green',label="testing data")
    plt.legend()

    if predictions is not None:
        plt.scatter(test_data,predictions,c='red',label='predictions')
        plt.legend()
    
```

```python
plot_predictions()
```

## Building up the pytorch model
* start with the random values (weight and bias)
* look at training values
* adjust the weights and get to the closer to ideal values
* By gradient descent and backpropagation

```python
# create a linear regression model class
class LinearRegressionModel(nn.Module):

    #inherits from nn.module
    #al most everything from pytorch model

    #modules could contain other modules
    def __init__(self):
        super().__init__()
        self.weights=nn.Parameter(torch.randn(1,requires_grad=True,dtype=torch.float32))

        self.bias=nn.Parameter(torch.randn(1,requirs_grad=True,dtype=torch.float32))

    def forword(self,x:torch.Tensor)->torch.Tensor:
        return self.weights*X+self.bias
        
```
