```python
weight=0.7
bias=0.3
```

```python
x=torch.arange(0,1,0.02).unsqueeze(dim=1)
```

```python
y=weight*x+bias
```

```python
plt.figure(dpi=1000)
plt.scatter(x,y,s=4)
```

```python
x_train=x[:int(0.8*len(x))]
y_train=y[:int(0.8*len(y))]
x_test=x[int(0.8*len(x)):]
y_test=y[int(0.8*len(y)):]
```

```python
plt.figure(dpi=1000)
plt.scatter(x_train,y_train,label='trainig data',s=4)
plt.scatter(x_test,y_test,c='g',label='testing data',s=4)
plt.legend()
```

```python
## 6.2 building a lnear model in pytorch
```

```python
class linearregression(nn.Module):

    def __init__(self):
        super().__init__()
        self.w=nn.Parameter(torch.randn(1,requires_grad=True,dtype=torch.float32))
        self.b=nn.Parameter(torch.randn(1,requires_grad=True,dtype=torch.float32))
    def forward(self,x:torch.Tensor):

            return self.w*x + self.b
```

```python
torch.manual_seed(42)
model=linearregression()
```

```python
list(model.parameters())
```

```python
lossfun=nn.L1Loss()
```

```python
optimizer=torch.optim.SGD(params=model.parameters(),lr=0.001)
```

```python
epochs=3000
epoch_v=[]
loss_v=[]
for epoch in range(epochs):
    epoch_v.append(epoch)
    model.train()

    y_p=model(x_train)

    loss=lossfun(y_p,y_train)
    
    print(loss.item())
    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
```

```python
list(model.parameters())
```

```python
with torch.inference_mode():

    y_pred=model(x_test)
y_pred
```

```python
plt.figure(dpi=1000)
plt.scatter(x_train,y_train,s=4)
plt.scatter(x_test,y_test,s=4)
plt.scatter(x_test,y_pred,s=4)
```

```python
class regression(nn.Module):
    def __init__(self):
        super().__init__()
        ## use nn.Linear() for creating model parameters
        ## linear layer proabing layer,dense layer
        
        self.linear_layer=nn.Linear(in_features=1,
                                    out_features=1)
        ## applies linear transformation in the formulae
        ## y=w*x+b

    def forward(self,x:torch.Tensor)->torch.Tensor:
        return self.linear_layer(x)
        
        
```

```python
x_train,y_train
```

```python
torch.manual_seed(42)
model1=regression()
```

```python
model1.state_dict()
```

```python
### training code
## we need a loss function
## we need an optimizer
## training loop
## testing loop
```

```python
loss_fn=nn.L1Loss()
optimizer=torch.optim.SGD(params=model1.parameters(),lr=0.01)
## we can also use the Adam
```

```python
#lets write a training loop
x_train.shape
```

```python
torch.manual_seed(42)
epochs=500
#put data device agnostic code for data
x_train=x_train.to(device)
y_train=y_train.to(device)
x_test=x_test.to(device)
y_test=y_test.to(device)
for epoch in range(epochs):
    model1.train()
    y_p=model1(x_train)
    loss=loss_fn(y_p,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    model1.eval()
    with torch.inference_mode():
        y_pred=model1(x_test)
        testloss=loss_fn(y_pred,y_test)
    # print
    if(epoch%10==0):
        print(f"epoch:{epoch} loss:{loss} testloss:{testloss}")
```
