```python
class linear(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight=nn.Parameter(torch.rand(1,requires_grad=True,dtype=torch.float32))
        self.bias=nn.Parameter(torch.rand(1,requires_grad=True,dtype=torch.float32))
    def forward(self,x):
        return (self.weight*x)+self.bias
```

```python
class regression(nn.Module):

    def __init__(self):
        super().__init__()
        self.weight=nn.Parameter(torch.rand(2,requires_grad=True,dtype=torch.float32))
        self.bias=nn.Parameter(torch.rand(1,requires_grad=True,dtype=torch.float32))
    def forward(self,x,x1):
        return (self.weight[0]*x)+(self.weight[1]*x1)+self.bias
```

```python
model1=regression()
```

```python
loss_fn=nn.L1Loss()
```

```python
loss_fn
```

```python
optimizer=torch.optim.SGD(params=model1.parameters(),lr=0.01)
```

```python
epochs=10000
# epochs is an one loop through the data (hyper paramters set by us)

## loop through the data
for epoch in range(epochs):

    #set the model to training mode
    model1.train() # train mode sets all parameters gradients arequire gradients

    # forword pass
    y_pred=model1(X_train,x1)

    # calculate the loss
    loss=loss_fn(y_pred,Y_train)
    print(loss)
    # # turns off gradient tracking

    # optimizer

    optimizer.zero_grad()

    # backpropogation

    loss.backward()

    optimizer.step()
    model1.eval()
    # by default the optimizer changes will be acumulate in the loop we have to zer them
```
