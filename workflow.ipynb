## Building up the pytorch model
* start with the random values (weight and bias)
* look at training values
* adjust the weights and get to the closer to ideal values
* By gradient descent and backpropagation

```python
# create a linear regression model class
class LinearRegressionModel(nn.Module):

    #inherits from nn.module
    #al most everything from pytorch model

    #modules could contain other modules
    def __init__(self):
        super().__init__()
        self.weights=nn.Parameter(torch.randn(1,requires_grad=True,dtype=torch.float32))

        self.bias=nn.Parameter(torch.randn(1,requires_grad=True,dtype=torch.float32))

    def forward(self,X:torch.Tensor)->torch.Tensor:
        return self.weights*X+self.bias
        
```

## Build a training Loop in PYtorch
A. some thing we need in loops
* loop through data
* forward pass(this involves data moving through `forword()` also called propagation
* make predictions on data
* calculate the loss (comparede forword pas predictions)
* optimizer zero grad
* Loss backword through network to calculate the gradient with respect to loss(back propagation)
* optimizer step(gradient descent)

```python
epochs=200
# epochs is an one loop through the data (hyper paramters set by us)
epoach_count=[]
loss_values=[]
test_loss_values=[]
## loop through the data
for epoch in range(epochs):
    epoach_count.append(epoch)
    #set the model to training mode
    model_0.train() # train mode sets all parameters gradients arequire gradients

    # forword pass
    y_pred=model_0(X_train)

    # calculate the loss
    loss=loss_fn(y_pred,Y_train)
    print(loss)
    
    # # turns off gradient tracking

    # optimizer

    optimizer.zero_grad()

    # backpropogation

    loss.backward()

    optimizer.step()

    model_0.eval()## turns off the different settings for testing/evaluating

    # dropout/batchnorm for testing

    with torch.inference_mode():
        ## we are just evaluating the answer

        #1.forward pass
        #2.torch.no_grad() is older versions of code work same as inference_mode less optimized
        test_pred=model_0(X_test)

        test_loss=loss_fn(test_pred,Y_test)
        losses=loss_fn(y_pred,Y_train)
        loss_values.append(loss.numpy())
        test_loss_values.append(test_loss)
    print(f"test loss {test_loss}")



    # by default the optimizer changes will be acumulate in the loop we have to zer them 
```
