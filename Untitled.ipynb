```python
import torch
import pandas as pd
import numpy as np
from torch import nn
import matplotlib.pyplot as plt
```

```python
data=torch.arange(0,1,0.02)
data.view(len(data),1)
```

```python
data=data.unsqueeze(1)
```

```python
data
```

```python
class linearregression(nn.Module):

    def __init__(self):
        super().__init__()
        self.linear=nn.Linear(in_features=1,out_features=1)
    def forward(self,x):
        return self.linear(x)
        
```

```python
torch.manual_seed(42)
model=linearregression()
```

```python
list(model.parameters())
```

```python
x=data
y=50*x+200
```

```python
x,y
```

```python
x_train=x[:int(0.8*len(x))].to('mps')
y_train=y[:int(0.8*len(y))].to('mps')
x_test=x[int(0.8*len(x)):].to('mps')
y_test=y[int(0.8*len(y)):].to('mps')
loss_fn=nn.L1Loss()
optimizer=torch.optim.SGD(params=model.parameters(),lr=0.1)
model.to('mps')

x_train.device
```

```python
epochs=3000
loss_values=[]
test_loss_val=[]
for epoch in range(epochs):

    model.train()

    y_p=model(x_train)
    loss=loss_fn(y_p,y_train)
    print(loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    loss_values.append(loss.to('cpu').detach().numpy())
    model.eval()
```

```python
plt.figure(dpi=1000)
plt.plot(loss_values)
```
