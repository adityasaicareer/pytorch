```python
import torch
from torch import nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Linear(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight=nn.Parameter(torch.rand(1,requires_grad=True,dtype=torch.float32))
        self.bias=nn.Parameter(torch.rand(1,requires_grad=True,dtype=torch.float32))

    def forward(self,X):
        return self.weight*X+self.bias
        
```

```python
torch.manual_seed(42)
model=Linear()
```

```python
model.bias
```

```python
print(model.parameters)
```

```python
print(list(model.parameters()))
```

```python
x=torch.rand(100)
print(x)
y=0.7*x+0.3
print(y)
with torch.inference_mode():
    y_prod=model(x[int(0.8*len(x)):])
```

```python
plt.figure(figsize=(10,10),dpi=1000)
plt.scatter(x[:int(0.8*len(x))],y[:int(0.8*len(y))],s=3)
plt.scatter(x[int(0.8*len(x)):],y[int(0.8*len(x)):],s=3)
```

```python
with torch.inference_mode():

    y_pred=model(x)
print(y_pred)
print(y)
```
